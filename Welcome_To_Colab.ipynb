```
gpa = float(input("Nhập GPA của sinh viên (0 - 10): "))
if gpa >= 8.5 and gpa <= 10:
  print("Học lực: Xuất sắc")
elif gpa >= 7.0:
  print("Học lực: Giỏi")
elif gpa >= 5.5:
  print("Học lực: Khá")
elif gpa >= 4.0:
  print("Học lực: Trung bình")
elif gpa >= 0:
  print("Học lực: Yếu")
else:
  print("GPA không hợp lệ")
```

```

n = int(input("Nhập số nguyên dương n: "))
tong = 0
for i in range(1, n + 1):
    if i % 2 == 0:   # Kiểm tra số chẵn
        tong += i
print("Tổng các số chẵn từ 1 đến", n, "là:", tong)
```

```
mat_khau_dung = "python123"
dem = 0
toi_da = 3
while dem < toi_da:
    mat_khau = input("Nhập mật khẩu: ")
    if mat_khau == mat_khau_dung:
        print("Đăng nhập thành công")
        break
    else:
        dem += 1
        print("Mật khẩu sai!")
if dem == toi_da:
    print("Tài khoản bị khóa")
```
